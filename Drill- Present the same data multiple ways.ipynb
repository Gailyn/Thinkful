```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
planes = pd.read_csv('plane_crashes_data.csv')
planes
```

```python
plt.hist(planes['month'])
# pros:  better than the box plot, ceasy to understand distribution and most common months 
#cons: hard to understand different distribution between closer months 
```

```python
planes['count'] = planes.month.value_counts()
#sns.barplot(x=planes['month'],y = planes['fatalities'] , palette = "pastel")

sns.factorplot(x="month", y="fatalities", data=planes, height=7, kind="bar",palette="rocket", ci=95)
#pros: easy comparison between months 
#cons: the grapg can get crowded, can simply see distribution of variable  
```

```python

sns.scatterplot(x=planes['month'], y=planes['fatalities'])
#Pro: can easily identify outliers
#con: cannot easily distinguish other cases to undersand more common cases 
```

```python
sns.boxplot(x=planes['month'], y=planes['fatalities'])
#pro: can undertand averages and ranges
#con: busy and hard to read
```

```python
mpg = pd.read_csv('mpg.csv')
mpg
```

```python
car_weight=mpg.loc[:,'weight']
car_hp=mpg.loc[:,'horsepower']
```

```python
sns.scatterplot(x=car_hp, y=car_weight)
#pro: easy to read, can examine outlier cases 
#con: hard to understand common distributions 
```

```python
sns.lineplot(x = car_weight, y=car_hp)
# pro: easy to understand general trend 
# con: busy and the amount of spikes are off putting 
```

```python
sns.jointplot(car_weight,car_hp, kind='kde')
#pro: can understand the general distribution of both variables and the relationship between the two, simple 
#con: hard to understand the effect of outliers in the kde plot
```

```python
# Challenge 

sns.set(style="whitegrid")
hatecrime = pd.read_csv("hatecrime.csv")

religion = hatecrime.loc[:, 'Religion']
so = hatecrime.loc[:, 'Sexual orientation']
race = hatecrime.loc[:,'Race']
total = religion + so + race 
hatecrime['total'] = total


masterplot = sns.PairGrid(hatecrime.sort_values("total", ascending=False),
                 x_vars=hatecrime.columns[3:6], y_vars=["State"],
                 height=10, aspect=.25)


masterplot.map(sns.stripplot, size=11, orient="h",
      palette="ch:s=1,r=-.1,h=1_r", linewidth=1, edgecolor="w")

masterplot.set(xlim=(0, 60), xlabel="number of incidents \n per jurisdiction", ylabel="")


titles = ["Racial Crimes", "Religion Crimes", "Sexual Orientation Crimes"]

for ax, title in zip(masterplot.axes.flat, titles):


    ax.set(title=title)

   
    ax.xaxis.grid(False)
    ax.yaxis.grid(True)

sns.despine(left=True, bottom=True)

#Racial hate crimes have a greater distribution than religious and sexual orientation oriented hate crimes 
```
