```python
# setting prediction target
# The goal is to predict arribution
Y = hr['Attrition']
hr = hr.drop('Attrition', 1)
Y = pd.get_dummies(Y)
hr = pd.get_dummies(hr)  
Y=Y['Yes']
```

```python
# Quick check of data types 
hr.dtypes
Y.dtypes
```

```python
# PCA for 4 components 
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score

# Use PCA to create new columns 
pca = PCA(n_components=4)
principalComponents = pca.fit_transform(hr)
pca_X = pd.DataFrame(data = principalComponents, 
                     columns = ['principal component 1', 'principal component 2', 'principal component 3', 'principal component 4'])



dt_start_time = time.time()
decision_tree = tree.DecisionTreeClassifier(
    criterion='entropy',
    max_features=4,
    max_depth=4,
)

decision_tree.fit(pca_X, Y)


# Fitting the model 
X = pca_X
decision_tree.fit(X, Y)

# Cross validation with 10% holdout group 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=10)

print('With 10% Holdout: ' + str(decision_tree.fit(X_train, y_train).score(X_test, y_test)))
print('Testing on Sample: ' + str(decision_tree.fit(X, Y).score(X, Y)))

# Cross validating using 10 folds  
from sklearn.model_selection import cross_val_score
print(cross_val_score(decision_tree, X, Y, cv=10))

dt_start_time = time.time()

print ("Decision Tree runtime: {}".format(time.time() - dt_start_time))


```

```python
from sklearn import ensemble
# random forest using PCA components
rf_start_time = time.time()
rfc = ensemble.RandomForestClassifier()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=10)
print('With 10% Holdout: ' + str(rfc.fit(X_train, y_train).score(X_test, y_test)))
print('Testing on Sample: ' + str(rfc.fit(X, Y).score(X, Y)))

# Cross validating using 10 folds  
from sklearn.model_selection import cross_val_score
print(cross_val_score(rfc, X, Y, cv=10))


print ("Random Forest runtime: {}".format(time.time() - rf_start_time))


```
