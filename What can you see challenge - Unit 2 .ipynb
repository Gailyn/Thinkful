```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
hatecrimes = pd.read_csv('hatecrime.csv')
hatecrimes
```

```python
race = hatecrimes.loc[:, 'Race']
plt.hist(race, bins= 200)
plt.title('Distribution of Racial hate crimes across America in 2013')
plt.show()
```

```python
religion = hatecrimes.loc[:, 'Religion']
plt.hist(religion, bins= 200, color='purple')
plt.title('Distribution of Religious hate crimes Across America in 2013 ')
plt.show()
```
