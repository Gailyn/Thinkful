```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
#My goal was to group states by region and compare the distribution of different crimes according to region

hatecrimes = pd.read_csv('hatecrime.csv')
hatecrimes
```

```python
race = hatecrimes.loc[:, 'Race']
plt.hist(race, density=True, bins= 200)
plt.title('Distribution of Racial hate crimes across America in 2013')
plt.show()
```

```python
religion = hatecrimes.loc[:, 'Religion']
plt.hist(religion, density=True, bins= 150, color='purple')
plt.title('Distribution of Religious hate crimes Across America in 2013 ')
plt.show()
```

```python
so = hatecrimes.loc[:, 'Sexual orientation']
plt.hist(so,density=True, bins= 200)
plt.title('Distribution of Sexual Orientation hate crimes across America in 2013')
plt.show()
```

```python
quarter1 = hatecrimes.loc[:, '1st quarter']
quarter1_squared = quarter1**2
plt.scatter(x = quarter1, y = quarter1_squared)
plt.title('Amount of incidences in the 1st Quarter of the year')
plt.show()

```

```python
disablity = hatecrimes.loc[:,'Disability']
plt.plot(disablity)
plt.title('Rate of Disablity Hate Crimes across the United States')
plt.show()


```
